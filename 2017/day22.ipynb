## Part One

```python
import numpy as np
```

```python
infmap = []

with open("day22_inp.dat", "r") as f:
    lines = f.readlines()
    center = (len(lines)-1)/2
    for i,line in enumerate(lines):
        for j,pos in enumerate(line.strip()):
            if pos == '#':
                infmap.append((j-center, center-i))
                
print(infmap)
```

```python
pos = np.array([0.,0.])
directions = np.array([[0,1], [1,0], [0,-1], [-1,0]])
dnow = 0
infs = 0
```

```python
def step(pos, dnow, infmap, infs):
    if tuple(pos) in infmap:
        dnow = (dnow+1)%4
        infmap.remove(tuple(pos))
    else:
        dnow = (dnow-1)%4
        infmap.append(tuple(pos))
        infs += 1
    pos = pos + directions[dnow]
    return pos, dnow, infmap, infs
```

```python
for i in range(10000):
    pos, dnow, infmap, infs = step(pos, dnow, infmap, infs)
```

## Part Two

```python
status = {}

with open("day22_inp.dat", "r") as f:
    lines = f.readlines()
    center = (len(lines)-1)/2
    for i,line in enumerate(lines):
        for j,pos in enumerate(line.strip()):
            if pos == '#':
                status[j-center, center-i] = 'I'
```

```python
status
```

```python
pos = np.array([0.,0.])
directions = np.array([[0,1], [1,0], [0,-1], [-1,0]])
dnow = 0
infs = 0
```

```python
def step(pos, dnow, status, infs):
    st = status.get(tuple(pos), 'C')
    if st == 'I':
        dnow = (dnow+1)%4
        status[tuple(pos)] = "F"
    elif st == 'W':
        status[tuple(pos)] = "I"
        infs += 1
    elif st == "F":
        dnow = (dnow+2)%4
        status[tuple(pos)] = "C"
    else:
        dnow = (dnow-1)%4
        status[tuple(pos)] = "W"
    pos = pos + directions[dnow]
    return pos, dnow, status, infs
```

```python
%%time
for i in range(10000000):
    pos, dnow, status, infs = step(pos, dnow, status, infs)
```
