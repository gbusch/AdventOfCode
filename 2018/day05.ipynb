# Day 5: Alchemical Reduction

https://adventofcode.com/2018/day/5

## Part One

```python
polymer = ""
polymer = [unit for unit in polymer]
```

```python
destroy = 1
while destroy > 0:
    destroy = 0
    i = 0
    while i>=0:
        if i >= (len(polymer)-1): break
        if (polymer[i].isupper() and polymer[i].lower()==polymer[i+1]) \
                or (polymer[i].islower() and polymer[i].upper()==polymer[i+1]): 
            del polymer[i+1]
            del polymer[i]
            destroy = 1
        else:
            i += 1
```

```python
print("Solution for Part One: {}.".format(len(polymer)))
```

## Part Two

```python
polymer = ""
polymer = [unit for unit in polymer]
```

```python
letters = {l.lower() for l in set(polymer)}
```

```python
minlen = 1e9
for l in letters:
    polrem = list(filter(lambda a: a!=l.lower() and a!=l.upper(), polymer))
    destroy = 1
    while destroy > 0:
        destroy = 0
        i = 0
        while i>=0:
            if i >= (len(polrem)-1): break
            if (polrem[i].isupper() and polrem[i].lower()==polrem[i+1]) \
                    or (polrem[i].islower() and polrem[i].upper()==polrem[i+1]): 
                del polrem[i+1]
                del polrem[i]
                destroy = 1
            else:
                i += 1
    print(l, len(polrem))
    if len(polrem) < minlen:
        minlen = len(polrem)
```

```python
print("Solution for Part Two: {}".format(minlen))
```
